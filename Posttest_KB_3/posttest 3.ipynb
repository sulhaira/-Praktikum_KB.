# Grafik Pie Chart

```python
data_pie = df['Status'].value_counts().rename_axis('status').reset_index(name='Id_count')
plt.figure(figsize=(10, 12))
colors = sns.color_palette('Set2', 8)
plt.pie(data_pie.Id_count, labels=data_pie.status, colors=colors, startangle=90, autopct='%.1f%%')
plt.title('status')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
