```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Life Expectancy Data.csv')
df.head()
```

# Grafik Pie Chart

```python
data_pie = df['Status'].value_counts().rename_axis('status').reset_index(name='Id_count')
plt.figure(figsize=(10, 12))
colors = sns.color_palette('Set2', 8)
plt.pie(data_pie.Id_count, labels=data_pie.status, colors=colors, startangle=90, autopct='%.1f%%')
plt.title('status')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

dari grafik pie di atas sudah jelas bahwa  Developing lebih banyak dari Developed dengan presentasi Developing 82.6% sedangkan Developed 17.4%

# Grafik Batang

```python
df_10 = df

fig = plt.figure(figsize = (15,8))
ax = fig.add_axes([0,0,1,1])
plt.title("Peminum Alcohol dari tahun ke tahun")
ax.set_xlabel("Year")
ax.set_ylabel("Alcohol")
ax.bar(df_10["Year"],df_10["Alcohol"], color = 'purple')
plt.show()
```

jumlah yang mengomsumsi alkohol dari tahub 2000-2003 jumlahnya hampir sama yaitu sekitar 13.0 hingga 14.0, dan terus mengalami peningkatan dari tahun 2004-2007 dan peminum alkohol terbanyak di tahun 2007 yaitu sekitar 17.5. kemudian 2008-2014 terus mengalami kenaikan dan penurunan secara berkala dan pada tahun 2015 mengalami penurunan yang sangat banyak hingga mencapai 10.0
